# Loading Data Set from excel file from multiple sheets for Sector wise and Removing default index column.

```python
#Loading file
url = 'C:\\Users\\Admin\\Desktop\\Data for Python Project EDA Nifty Sector Indices.xlsx'
sheetauto=pd.read_excel(url,sheet_name='NIFTYAUTO')
sheetpharma=pd.read_excel(url,sheet_name='NIFTYPHARMA')
sheetmetal=pd.read_excel(url,sheet_name='NIFTYMETAL')
sheetfmcg=pd.read_excel(url,sheet_name='NIFTYFMCG')
sheetbank=pd.read_excel(url,sheet_name='NIFTYBANK')
sheetit=pd.read_excel(url,sheet_name='NIFTYIT')
#Adding Year to each sector data
sheetauto['Year']= sheetauto['Date'].dt.year
sheetpharma['Year']= sheetpharma['Date'].dt.year
sheetmetal['Year']= sheetmetal['Date'].dt.year
sheetfmcg['Year']= sheetfmcg['Date'].dt.year
sheetbank['Year']= sheetbank['Date'].dt.year
sheetit['Year']= sheetit['Date'].dt.year

#Adding Year to each sector data
sheetauto['Month']= sheetauto['Date'].dt.month
sheetpharma['Month']= sheetpharma['Date'].dt.month
sheetmetal['Month']= sheetmetal['Date'].dt.month
sheetfmcg['Month']= sheetfmcg['Date'].dt.month
sheetbank['Month']= sheetbank['Date'].dt.month
sheetit['Month']= sheetit['Date'].dt.month

#Removing default index column
sheetauto = sheetauto.set_index(['Date'])
sheetpharma = sheetpharma.set_index(['Date'])
sheetmetal = sheetmetal.set_index(['Date'])
sheetfmcg = sheetfmcg.set_index(['Date'])
sheetbank = sheetbank.set_index(['Date'])
sheetit = sheetit.set_index(['Date'])
```

# Question11:Find the  Daily Percentage Change and Volatality for sectors

```python
#formula for percentage change = rt= (p1/pt-1)-1
#adding the column as return as per formula for all sectors

sheetauto['Returns']=(sheetauto['Close']/sheetauto['Close'].shift(1))-1
sheetpharma['Returns']=(sheetpharma['Close']/sheetpharma['Close'].shift(1))-1
sheetmetal['Returns']=(sheetmetal['Close']/sheetmetal['Close'].shift(1))-1
sheetfmcg['Returns']=(sheetfmcg['Close']/sheetfmcg['Close'].shift(1))-1
sheetbank['Returns']=(sheetbank['Close']/sheetbank['Close'].shift(1))-1
sheetit['Returns']=(sheetit['Close']/sheetit['Close'].shift(1))-1

#just shwoing the out only for NIFTY AUTO
sheetauto.head()
```

```python
#VOLATALITY
plt.figure(figsize=(15,8))
sheetauto['Returns'].hist(bins=100,label='NIFTYAUTO',alpha=0.5)
sheetpharma['Returns'].hist(bins=100,label='NIFTYPHARMA',alpha=0.5)
sheetmetal['Returns'].hist(bins=100,label='NIFTYMETAL',alpha=0.5)
sheetfmcg['Returns'].hist(bins=100,label='NIFTYFMCG',alpha=0.5)
sheetbank['Returns'].hist(bins=100,label='NIFTYBANK',alpha=0.5)
sheetit['Returns'].hist(bins=100,label='NIFTYIT',alpha=0.5)
plt.legend()
plt.show
```
